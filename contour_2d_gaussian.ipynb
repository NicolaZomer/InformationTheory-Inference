```python
import matplotlib.pyplot as plt
import numpy as np
from   scipy.stats import multivariate_normal

N    = 1000
X    = np.linspace(-0.025, 0.025, N)
Y    = np.linspace(-0.025, 0.025, N)
X, Y = np.meshgrid(X, Y)
pos  = np.dstack((X, Y))
rv   = multivariate_normal([0, 0], 0.0001*np.eye(N=2))
Z    = rv.pdf(pos)

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, CS.levels, inline=True, fontsize=10)

plt.grid(alpha=0.5)
plt.show()
```
